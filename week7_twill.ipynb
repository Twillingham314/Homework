```SQL
SELECT stationId, count(*) as usagecount 
FROM EVCharging
GROUP By stationId 
ORDER by usagecount DESC
LIMIT 5
```

3. Question 3: Using dimfacility and factCharge, write a query to find out which type of facility (GROUP BY) has the most amount of charging stations. Return type Facility and name the calculated column numStation. Order the result set from highest to lowest number of charging stations.
```SQL
SELECT b.typeFacility, COUNT(DISTINCT a.stationId) AS numStation
FROM factCharge a
INNER JOIN dimFacility b 
On a.facilityID = b.FacilityKey
GROUP by b.typeFacility
ORDER by numStation DESC;
```
